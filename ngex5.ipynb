AIM #2: Finding issues (empty, NAs, incorrect value, incorrect format, outliers, etc.) 
1. Find out how many missing values there are in the dataset
2. For the 'Age' column, find the best way to handle the missing values
    2.1. Use an appropriate plot to study the nature of the 'Age' column
    2.2. Figure out what is the best way to calculate the central tendency of the 'Age' column based on the above plot
    2.3. Using the most suitable central tendency measure, fill the missing values in the age column
3. Decide what is the best way to handle the missing values in the 'Cabin' columns
4. Similarly, decide what is the best way to handle the missing values in the 'Embarked' columns
5. Handle the incorrect data under the 'Survived' columns using appropriate measure
6. Handle the incorrectly formatted data under the 'Fare' column


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('titanic.csv')

missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)

sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

median_age = df['Age'].median()
print("Median Age:", median_age)

df['Age'].fillna(median_age, inplace=True)

df.drop(columns=['Cabin'], inplace=True)

mode_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(mode_embarked, inplace=True)

df['Survived'] = pd.to_numeric(df['Survived'], errors='coerce').fillna(0).astype(int)

df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')
median_fare = df['Fare'].median()
df['Fare'].fillna(median_fare, inplace=True)

print("\nData after cleaning:")
print(df.head())
```

AIM #3: Grouping 
1. Find out the average fare grouped by Pclass
    1.1. Plot the above using a suitable plot
2. Find out the average fare grouped by Sex
    2.1. Plot the above using a suitable plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('titanic.csv')


df['Fare'] = df['Fare'].replace('[\$,]', '', regex=True).astype(float)


avg_fare_by_pclass = df.groupby('Pclass')['Fare'].mean()


plt.figure(figsize=(10, 6))
sns.barplot(x=avg_fare_by_pclass.index, y=avg_fare_by_pclass.values)
plt.xlabel('Pclass')
plt.ylabel('Average Fare')
plt.title('Average Fare by Pclass')
plt.show()


avg_fare_by_sex = df.groupby('Sex')['Fare'].mean()


plt.figure(figsize=(6, 4))
sns.barplot(x=avg_fare_by_sex.index, y=avg_fare_by_sex.values)
plt.xlabel('Sex')
plt.ylabel('Average Fare')
plt.title('Average Fare by Sex')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('titanic.csv')

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Fare', bins=20, kde=True, color='lightgreen', edgecolor='black')
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Pclass', palette='Set3')
plt.title('Distribution of Pclass')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Survived', palette='pastel')
plt.title('Distribution of Survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Embarked', palette='Set2')
plt.title('Distribution of Embarked')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Fare', hue='Survived', bins=20, kde=True, palette='Set1', edgecolor='black')
plt.title('Distribution of Fare grouped by Survived')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.legend(title='Survived', loc='upper right')
plt.grid(axis='y', alpha=0.75)
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Fare', hue='Pclass', bins=20, kde=True, palette='Set2', edgecolor='black')
plt.title('Distribution of Fare grouped by Pclass')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.legend(title='Pclass', loc='upper right')
plt.grid(axis='y', alpha=0.75)
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Survived', bins=20, kde=True, palette='Set3', edgecolor='black')
plt.title('Distribution of Age grouped by Survived')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(title='Survived', loc='upper right')
plt.grid(axis='y', alpha=0.75)
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Pclass', bins=20, kde=True, palette='Set1', edgecolor='black')
plt.title('Distribution of Age grouped by Pclass')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(title='Pclass', loc='upper right')
plt.grid(axis='y', alpha=0.75)
plt.show()

df['Family_Size'] = df['SibSp'] + df['Parch']
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Family_Size', hue='Survived', palette='Set2')
plt.title('Distribution of Family Size grouped by Survived')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Family_Size', hue='Pclass', palette='Set3')
plt.title('Distribution of Family Size grouped by Pclass')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.legend(title='Pclass', loc='upper right')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='Set1')
plt.title('Relationship between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', loc='upper right')
plt.grid(True)
plt.show()
```

AIM #5: Correlation

2. Find correlation between 'Age' and 'Fare'
